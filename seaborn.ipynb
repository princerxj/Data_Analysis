## Data Visualization with Seaborn 

Seaborn is a Python Visualization Library based on Matplotlib that provides a high-level interface for drawing attractive and informative statistical Graphics. Seaborn helps in creating complex visualization with just a few lines of code. Here , we will cover the basics of Seaborn, including the creation of various types of plots and customizing them

```python
import seaborn as sns
```

```python
## Basic Plotting with Seaborn
tips = sns.load_dataset('tips')
tips
```

```python
## Creater a scatter 
import matplotlib.pyplot as plt 

sns.scatterplot(x = 'total_bill', y = 'tip', data = tips)
plt.title("Scatter plot of total bill vs tip")
plt.show()
```

```python
## Line Plot 
sns.lineplot(x = 'size', y = 'total_bill', data = tips)
plt.title("Line Plot of the Total bill vs size")
```

```python
## Categorical Plots 
## Bar Plot
sns.barplot(x = 'day', y = 'total_bill', data=tips)
plt.title('Bar Plot of Total Bill vs Day')
```

```python
## Box Plot 
sns.boxplot(x = 'day', y = 'total_bill', data = tips)
```

```python
## Violin Plot 

sns.violinplot(x = 'day', y = 'total_bill', data = tips)
```

```python
## Histograms 
sns.histplot(tips['total_bill'], bins = 10, kde = True)
```

```python
## KDE Plot - (Kernel Density Plot)
sns.kdeplot(tips['total_bill'], fill= True)
```

```python
## Pair Plot 
sns.pairplot(tips)
```

```python
## HeatMap 
corr = tips[['total_bill', 'tip', 'size']].corr()
corr
```

```python
sns.heatmap(corr, annot=True, cmap='coolwarm')
```

```python
import pandas as pd 
sales_df = pd.read_csv('sales_data.csv')
sales_df.head(5)
```

```python
## Total Sales by Product Plot 
plt.figure(figsize=(10,6))
sns.barplot(x = 'Product Category', y = 'Total Revenue', data = sales_df, estimator=sum)
plt.xlabel('Product')
plt.ylabel('Sales')
plt.show()
```

```python
## Total Sales by Region Plot 
plt.figure(figsize=(10,6))
sns.barplot(x = 'Region', y = 'Total Revenue', data = sales_df, estimator=sum)
plt.title('Total Sales vs Region')
plt.xlabel('Region')
plt.ylabel('Sales')
plt.show()
```
