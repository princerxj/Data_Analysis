```python
import requests
import pandas as pd
import time

contest_id = 2178
base_url = "https://codeforces.com/api/contest.standings"

all_records = []

start = 1
batch_size = 500 

while True:
    params = {
        "contestId": contest_id,
        "from": start,
        "count": batch_size
    }

    response = requests.get(base_url, params=params).json()

    if response["status"] != "OK":
        break

    rows = response["result"]["rows"]
    if not rows:   # no more data
        break

    for r in rows:
        all_records.append({
            "rank": r["rank"],
            "handle": r["party"]["members"][0]["handle"],
            "points": r["points"],
            "penalty": r["penalty"]
        })

    start += batch_size
    time.sleep(0.5)

df = pd.DataFrame(all_records)
df.head()
```

```python
df.to_csv("rankings.csv")
```

```python
import csv

handle = "princerxj"
def get_rank(csv_path, handle):
    with open(csv_path, newline='', encoding='utf-8') as file :
        reader = csv.DictReader(file)
        for row in reader :
            if row["handle"].lower() == handle.lower():
                return int(row["rank"])
    return None

rank = get_rank("rankings.csv", handle)
print(f"{handle} got a rank of {rank} in CF Contest 2178")
```
