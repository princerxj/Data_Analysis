## Pandas - DataFrame and Series
Pandas is a powerful data manipulation library in Pyhton, widely used for data analysis and data cleaning. It provides two primary data structures: Series and DataFrame. A series is one-dimensional array-like object, while a DataFrame is a two-dimensional, size-mutable, and potentially heterogeneous tabular data structure with labeled axes(rows and columns).

```python
import pandas as pd
```

```python
# Series

import pandas as pd
data = [1,2,3,4,5]
series = pd.Series(data)
print("Series : \n", series)
print(type(series))
```

```python
## Create a series from dictionary elements
data = {'a':1, 'b':2, 'c':3}
series_dict = pd.Series(data)
print(series_dict)
```

```python
data = [10,20,30]
index = ['a', 'b', 'c']

pd.Series(data, index)
```

```python
## DataFrame
## Create a DataFrame from a dictionary of lists

data = {
    'Name' : ["Krish", "John", "Jack"],
    'Age' : [25,30,45],
    'City' : ["Bangalore", "New York", "Chichago"]
}

df = pd.DataFrame(data)
print(df)
print(type(df))
```

```python
import numpy as np 
np.array(df)
```

```python
## Create a dataFrame from a list of dictionary 
data = [
    {'Name' : "Krish", 'Age' : 32, 'City' : "Bangalore"},
    {'Name' : "John", 'Age' : 22, 'City' : "Delhi"}, 
    {'Name' : "Jack", 'Age' : 72, 'City' : "Florida"}, 
    {'Name' : "Elon", 'Age' : 68, 'City' : "Mars"}
]

df = pd.DataFrame(data)
print(df)
print(type(df))
```
